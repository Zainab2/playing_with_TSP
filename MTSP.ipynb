```python
#This cell contains all the main functions .
#This cell only for complete graphs .
#Later cell will be used to discuss the case of  incomplete graphs with christofides and dynamic programming .
import networkx as nx
import os
import matplotlib.pyplot as plt
import math
from itertools import permutations
from itertools import combinations
from itertools import chain
import numpy as np
from networkx.algorithms.matching import max_weight_matching
from networkx.algorithms.euler import eulerian_circuit
import pandas as pd
import time

def load_coordinates_from_file(filename) :
    """This function takes the name of the file that containes the coordinates and returns list of points.  """
    """filename : String"""
    with open(filename,'r') as file  :
        list_of_points = [] 
        points = file.readlines()
        for point in points :
            point = point.strip().split(" ")
            if(point[0] == 'EOF') :
                return list_of_points
            x_coordinate = point[1]
            y_coordinate = point[2]
            list_of_points.append((x_coordinate,y_coordinate))
            
def load_graph(filename) :
    """
    This functions is used to create graph instance from specific file with specific format .
    File foramt :
    vertex vertex weight
    .
    .
    .
    .
    .
    EOF
    """
    file = open(filename,'r')
    lines = file.readlines()
    graph = nx.Graph()
    for line in lines :
        line = line.strip().split(" ")
        if (line[0] == 'EOF') :
            return graph
        graph.add_edge(int(line[0]),int(line[1]),weight = int(line[2]))
        
def generate_tight_example(number_of_vertices,file = False) :
    """
    This function is used to generate graph instance which has a special case that called tight example .
    If the file parameter is True ,this means generate file representing tight example  .
    """
    graph = nx.Graph()
    for i in range(number_of_vertices) :
        for j in range(i) :
            if(j==0) :
                graph.add_edge(i,j,weight = 1)
            else :
                graph.add_edge(i,j,weight = 2)
    if(file == True) :
        file = open("C:/Users/Ayman/datasets/tightExample"+str(graph.number_of_nodes())+".txt",'w')
        for edge in graph.edges() :
            file.write(str(edge[0])+" "+str(edge[1])+" "+str(graph[edge[0]][edge[1]]['weight'])+"\n")
        file.write('EOF')
        file.close()
        return graph
        
def compute_distance(point1,point2) :
    """this function takes two points,each of them is 2-tuple and returns a Euclidean distance between them.  """
    """point1 : 2-tuple"""
    """point2 : 2-tuple"""
    x1 = float(point1[0])
    y1 = float(point1[1])
    x2 = float(point2[0])
    y2 = float(point2[1])
    distance  = int(math.sqrt((x1 - x2) **2 + (y1 - y2) **2) + 0.5)
    return distance

def convert_coordinates_to_graph(list_of_points) :
    """This function takes a list of 2-tuples and returns a graph instance of networkx' class.  """
    """list_of_points : list of 2-tuples"""
    graph = nx.Graph()
    number_of_vertices = len(list_of_points)
    for i in range(number_of_vertices) :
        for j in range(i) :
            graph.add_edge(i,j,weight = compute_distance(list_of_points[i],list_of_points[j]))
    return graph

def plot_cycle(list_of_points, cycle):
    """This function takes a list_of_points that represents the coordinates of the points and a cycle which is a list 
    and draw the cycle.    """
    """list_of_points : list of 2-tuples"""
    x_coordinates = [float(list_of_points[i][0]) for i in cycle]
    y_coordinates = [float(list_of_points[i][1]) for i in cycle]
    x_coordinates.append(float(list_of_points[cycle[0]][0]))
    y_coordinates.append(float(list_of_points[cycle[0]][1]))
    plt.plot(x_coordinates, y_coordinates, 'or-', )
    plt.show()

def plot_graph(graph) :
    """
    This function takes a graph instance and draw it .
    Note : Here positions of vertices is not considered .
    """
    pos = nx.spring_layout(graph)
    nx.draw_networkx_nodes(graph,pos)
    nx.draw_networkx_edges(graph,pos)
    nx.draw_networkx_labels(graph,pos)
    #nx.draw_networkx_edge_labels(graph,pos)
    plt.show()
    

def draw_graph(filename) :
    """
    This function takes positions of the vertices into account .
    """
    list_of_points = load_coordinates_from_file(filename)
    graph = nx.Graph()
    for index , point in enumerate(list_of_points) :
        graph.add_node(index,pos = (float(point[0]),float(point[1])))
    for i in range(len(list_of_points)) :
        for j in range(i) :
            graph.add_edge(i,j,weight = int(compute_distance(list_of_points[i],list_of_points[j])))
    pos = nx.get_node_attributes(graph,'pos')
    nx.draw_networkx_nodes(graph,pos)
    nx.draw_networkx_edges(graph,pos)
    nx.draw_networkx_labels(graph,pos)
    
def compute_cycle_length(graph,cycle) :
    """This function takes a graph instance and a cycle and returns the length of this cycle.  """
    """graph : networkx instance"""
    """cycle : tuple and in some cases can be passed as a list. """
    length = 0
    for vertex_id in range(len(cycle)) :
        length += graph[cycle[vertex_id]][cycle[(vertex_id + 1) % len(cycle)]]['weight']
    return length
  
    
def brute_force(graph) :
    """This function takes a graph instance and computes a minimum weigth Hamiltonian cycle on it.  """
    """graph : networkx instance"""
    number_of_vertices = graph.number_of_nodes()
    minimum_weight = float("+inf")
    minimum_weight_hamiltonian_cycle = None
    #1.iterate through all permutations and choose the best cycle (here : cycle with the minimum length)
    for cycle in permutations(range(number_of_vertices)) :
        #This line is added in 2/6/2020 10:08 pm
        #The reason is to reduce the number of candidate cycles from n! to (n-1)!
        if(cycle[0] != 0 ) :
            return minimum_weight_hamiltonian_cycle , minimum_weight
        current_weight = compute_cycle_length(graph,cycle)
        if(current_weight < minimum_weight ) :
            minimum_weight = current_weight
            minimum_weight_hamiltonian_cycle = cycle
    return minimum_weight_hamiltonian_cycle , minimum_weight
            

def compute_lower_bound(graph,cycle) :
    """This function takes a graph instance and a subcycle and returns the lower bound in this stage.  """
    """graph : networkx instance"""
    """cycle : tuple and in some cases can be passed as a list. """
    """This function is used in the function that implements branch and bound algorithm.  """
    #1.computing the length of the path of visited nodes .
    weight_of_visited_vertices = 0
    for i in range(len(cycle) - 1 ) :
        weight_of_visited_vertices += graph[cycle[i]][cycle[i+1]]['weight']
    #2.computing the remaining graph .
    unvisited = [vertex for vertex in graph.nodes() if vertex not in cycle]
    remaining_graph = graph.subgraph(unvisited)
    #3.computing the weight of MST in the remaining graph .
    mst_edges = nx.minimum_spanning_edges(remaining_graph)
    mst_weight = 0 
    for edge in mst_edges :
        mst_weight += edge[2]['weight']
    #4.Defining the behaviour when we reach the end of the cycle .
    #Note : In this implementation the length of cycle can not be equal to zero .
    #Case when all vertices either in cycle or in the remaining graph .
    if (len(cycle) == 0 or len(cycle) == graph.number_of_nodes()) :
        return mst_weight + weight_of_visited_vertices
    #5.handling the case when some vertices belong to the remaining graph and others belong to the cycle .
    first_vertex_in_cycle = cycle[0]
    last_vertex_in_cycle = cycle[-1]
    min_weight_to_first_vertex = float("inf")
    min_weight_from_last_vertex = float("inf")
    for vertex in graph.nodes() :
        if vertex not in cycle :
            if(graph[vertex][first_vertex_in_cycle]['weight'] < min_weight_to_first_vertex) :
                min_weight_to_first_vertex = graph[vertex][first_vertex_in_cycle]['weight']
    for vertex in graph.nodes() :
        if vertex not in cycle :
            if(graph[last_vertex_in_cycle][vertex]['weight'] < min_weight_from_last_vertex) :
                min_weight_from_last_vertex = graph[last_vertex_in_cycle][vertex]['weight']
    return weight_of_visited_vertices + mst_weight + min_weight_to_first_vertex + min_weight_from_last_vertex

def branch_and_bound_helper(graph,list_of_cycles,cycle=None,current_min = float("inf") ) :
    """
    This function takes three parameters :
    graph : networkx's instance in which we want to find a minimum weight Hamiltonian cycle .
    cycle : The vertex that is a part of the cycle under cosideration at each stage (tuple) .
    current_min : The weight of the best solution at each stage .
    This function is an implementation of brunch and bound algoritm .
    This function returns the minimum weight Hamitonian cycle and its weight .
    
    """
    #1.If there is no root . we define the root of search tree .
    if cycle is None :
        cycle = [0]
    #2.The case when we reach the end of the cycle and we're ready to compute its weight.
    if (len(cycle) == graph.number_of_nodes()) :
        weight = sum([graph[cycle[i]][cycle[(i + 1) % len(cycle)]]['weight'] for i in range(len(cycle))])
        return weight
    #3.Determining the unvisited vertices .
    unvisited = []
    for vertex in graph.nodes():
        if vertex not in cycle:
            unvisited.append((graph[cycle[-1]][vertex]['weight'], vertex))
    #4.Fininding nearest neighbors to current last vertex of the current part of a cycle .
    unvisited = sorted(unvisited)
    #5.Trying reach the end of the cycle .
    for (weight,vertex) in unvisited :
        #Tuples doesn't support append .
        modified_cycle = list(cycle)
        modified_cycle.append(vertex)
        #To check if we by adding this vertex to the current path(part of cycle) able to reach a better solution
        #better solution -----> There is a possiblity to find smaller weight cycle by proceeding in this path .
        if compute_lower_bound(graph,modified_cycle) < current_min :
            new_weight = branch_and_bound_helper(graph,list_of_cycles,modified_cycle,current_min)
            #If The newly found cycle is better than the previously discovered then save it .
            if new_weight < current_min :
                current_min = new_weight
                if(len(modified_cycle) == graph.number_of_nodes()) :
                    list_of_cycles.append(modified_cycle)
    return current_min

def branch_and_bound(graph) :
    list_of_cycles = list()
    weight_of_best_hamiltonian_cycle = branch_and_bound_helper(graph,list_of_cycles)
    return tuple(list_of_cycles[-1]) , weight_of_best_hamiltonian_cycle

def dynamic_programming(graph) :
    """This function takes only one parameter
    graph : networkx's instance .
    This is the implementation of Dynamic Programming approach.
    This function returns the minimum weight Hamiltonian cycle wth its weight .
    """
    #1.We must create a list that contains all possible subsets of our {all vertices without zero }set .
    n = graph.number_of_nodes()
    chain_of_subsets = chain.from_iterable(combinations(range(1,n),r) for r in range(n+1))
    #2.Handle the case of subsets with only one element ---> this means edges from 0 to each other vertex .
    weight_of_subset = dict()
    for vertex in range(1,n) :
        weight_of_subset[(vertex,),vertex] = graph[0][vertex]['weight']
    #3.Handling the case of subsets that have more than one element .
    for subset in chain_of_subsets :
        if(len(subset) > 1) :
            for vertex in subset :
                sub_subset = [node for node in subset if node != vertex]
                sub_subset = tuple(sub_subset)
                weight_of_subset[subset,vertex] = min([weight_of_subset[sub_subset,extension] + graph[extension][vertex]['weight'] for extension in sub_subset])
    #4.Preparing the results to be returned .
    min_weight = float("inf")
    cycle = tuple()
    new_last_vertex = None
    for vertex in range(1,n) :
        if(weight_of_subset[tuple(range(1,n)),vertex] + graph[vertex][0]['weight'] < min_weight) :
            min_weight = weight_of_subset[tuple(range(1,n)),vertex] + graph[vertex][0]['weight']
            new_last_vertex = vertex
    new_weight = min_weight  - graph[new_last_vertex][0]['weight']
    path = list()
    path.append(0)
    path.append(new_last_vertex)
    for i in reversed(range(1,n-1)) :
        list_of_combinations = list(combinations(range(1,n),i))
        for j in range(n) :
            for sub in list_of_combinations :
                if(j in sub and set(path).isdisjoint(sub)) :
                    if(weight_of_subset[sub,j] + graph[j][new_last_vertex]['weight'] - new_weight < 0.0000000001) :
                        new_weight = new_weight - graph[j][new_last_vertex]['weight']
                        new_last_vertex = j
                        path.append(j)
    return path , min_weight

def double_approximation(graph,source_vertex = 0) :
    """This function implements the double approximation algorithm to solve Metric TSP."""
    #1.Compute the MST for the inputed graph .
    mst = nx.minimum_spanning_tree(graph) 
    #2.Traverse the mst .
    cycle = nx.dfs_preorder_nodes(mst,source_vertex)
    cycle = tuple(cycle)
    weight = compute_cycle_length(graph,cycle)
    return cycle , weight

def modified_double_approximation(graph) :
    """
    This function is cosidered as a modification for double approximation function .
    The mian point here is that the order of visiting the vertices is important because it plays a major role in identifying
    the bypasses .
    """
    min_weight = float("inf")
    min_cycle = None 
    n = graph.number_of_nodes()
    for vertex in range(n) :
        cycle , weight = double_approximation(graph,vertex)
        if(weight < min_weight) :
            min_weight = weight
            min_cycle = cycle
    return min_cycle , min_weight

def random_permutation(graph,threshold = 2,max_time = float("inf") ) :
    tic = time.time()
    weight = nx.minimum_spanning_tree(graph).size(weight = 'weight')
    min_weight = float("inf")
    min_cycle = None
    number_of_vertices = graph.number_of_nodes()
    toc = time.time()
    while(round((toc - tic) * 1000) < max_time and min_weight > threshold * weight) :
        cycle = np.random.permutation(number_of_vertices)
        min_weight = compute_cycle_length(graph,cycle)
        toc = time.time()
    if(cycle is None) :
        return None , None
    else :
        return tuple(cycle) , min_weight
    
def nearest_neighbor(graph,source_vertex = 0) :
    """
    This function is the implementation of choosing the nearest vertex(neighbor) each time .
    graph : networkx's instance .
    source_vertex : The first vertex in the cycle .
    """
    #1.Intialization step for neccessary variables .
    number_of_vertices = graph.number_of_nodes()
    cycle = list([source_vertex])
    last_visited_vertex = source_vertex
    unvisited = list([vertex for vertex in range(number_of_vertices) if vertex != source_vertex])
    #2.Passing throught all unvisited vertices .
    for counter in range(number_of_vertices - 1):
        next_visited_vertex = None
        weight = float('inf')
        for vertex in unvisited :
            if(graph[last_visited_vertex][vertex]['weight'] < weight) :
                next_visited_vertex = vertex
                weight = graph[last_visited_vertex][vertex]['weight']
        cycle.append(next_visited_vertex)
        unvisited.remove(next_visited_vertex)
        last_visited_vertex = next_visited_vertex
    weight = compute_cycle_length(graph,cycle)
    return tuple(cycle) , weight

def modified_nearest_neighbor(graph) :
    """
    This function is slight modification of nearest neighbor ,considering that the starting vertex can affect the result .
    Main point : Order is important (Each vertex affects the choosing of next one. )
    """
    number_of_vertices = graph.number_of_nodes()
    cycle = None
    weight = float("inf")
    for vertex in range(number_of_vertices) :
        new_cycle , new_weight = nearest_neighbor(graph,vertex)
        if (new_weight < weight) :
            weight = new_weight
            cycle = new_cycle
    return cycle , weight
    
def christofides(graph,source_vertex = 0) :
    """
    This function is an implementation of Christofides algorithm .
    graph : networkx instance .
    source_vertex : The starting and ending point of the cycle .
    """
    #1.compute MST .
    mst = nx.minimum_spanning_tree(graph)
    #2.find the vertices with odd degrees .
    odd_degree_vertices = list()
    for vertex in graph.nodes() :
        if(mst.degree(vertex) % 2 != 0) :
            odd_degree_vertices.append(vertex)
    #3.build subgraph with only vertices with odd degrees and the edges between them .
    odd_vertices_graph = graph.subgraph(odd_degree_vertices)
    #4.compute minimum weight perfect matching on the odd vertices subgraph .
    for edge in odd_vertices_graph.edges() :
        odd_vertices_graph[edge[0]][edge[1]]['weight'] = -1 * odd_vertices_graph[edge[0]][edge[1]]['weight']
    minimum_weight_perfect_matching = list(max_weight_matching(odd_vertices_graph, maxcardinality=True))
    #To correct negative edges .
    for edge in graph.edges() :
        graph[edge[0]][edge[1]]['weight'] = abs(graph[edge[0]][edge[1]]['weight'])
    #5.combine the mst with the edges from minimum weight perfect matching .
    multi_graph = nx.MultiGraph(mst)
    for edge in minimum_weight_perfect_matching :
        multi_graph.add_edge(edge[0],edge[1],weight = graph[edge[0]][edge[1]]['weight'])
    #6.find Euler's cycle
    euler_cycle = list(eulerian_circuit(multi_graph,source = source_vertex))
    cycle = list()
    for edge in euler_cycle :
        if(edge[0] not in cycle) :
            cycle.append(edge[0])
        if(edge[1] not in cycle) :
            cycle.append(edge[1])
    weight = compute_cycle_length(graph,tuple(cycle))
    return tuple(cycle) , weight

def modified_christofides(graph) :
    """
    This function is used to negotiate the effects of visiting order .
    """
    min_weight = float("inf")
    min_cycle = None
    for vertex in range(graph.number_of_nodes()) :
        cycle , weight = christofides(graph,vertex)
        if (weight < min_weight ) :
            min_weight = weight
            min_cycle = cycle
    return min_cycle , min_weight
      
def first_choice_hill_climbing(graph,ratio = 100) :
    """
    This functions is an implementatioan of local search algorithm called first choice hill climbing .
    There are many reasons that lead to choose this version among other hill climbing versions .
    Note : Here we consider only change between two edges .
    """
    #1.produce a random permutation .
    number_of_vertices = graph.number_of_nodes()
    cycle = np.random.permutation(number_of_vertices)
    counter = np.random.randint(0,10)
    for i in range(counter) :
        np.random.shuffle(cycle)
    weight = compute_cycle_length(graph,cycle)
    #2.change the positions of two randomly choosen vertices in the cycle if the condition holds .
    for iteration in range(ratio * number_of_vertices) :
        num1 = np.random.randint(1,number_of_vertices)
        num2 = np.random.randint(1,number_of_vertices)
        while(num1 == num2) :
            num1 = np.random.randint(1,number_of_vertices)
            num2 = np.random.randint(1,number_of_vertices)
        first_index = min(num1,num2)
        second_index = max(num1,num2)
        temp_cycle = cycle.copy()
        temp_cycle[first_index:second_index + 1] = list(reversed(temp_cycle[first_index:second_index + 1]))
        new_weight = compute_cycle_length(graph,temp_cycle)
        if(new_weight < weight) :
            weight = new_weight
            cycle = temp_cycle
    return tuple(cycle) , weight 

def christofides_first_choice_hill_climbing(graph,ratio = 100) :
    """
    This is a modification of first choice hill climbing .
    instead of starting with a random cycle , we start with a cycle given as a solution by christofides algorithm .
    """
    #1.produce a christofides solution .
    number_of_vertices = graph.number_of_nodes()
    cycle , weight = christofides(graph)
    cycle = list(cycle)
    #2.change the positions of two randomly choosen vertices in the cycle if the condition holds .
    for iteration in range(ratio * number_of_vertices) :
        num1 = np.random.randint(1,number_of_vertices)
        num2 = np.random.randint(1,number_of_vertices)
        while(num1 == num2) :
            num1 = np.random.randint(1,number_of_vertices)
            num2 = np.random.randint(1,number_of_vertices)
        first_index = min(num1,num2)
        second_index = max(num1,num2)
        temp_cycle = cycle.copy()
        temp_cycle[first_index:second_index + 1] = list(reversed(temp_cycle[first_index:second_index + 1]))
        new_weight = compute_cycle_length(graph,temp_cycle)
        if(new_weight < weight) :
            weight = new_weight
            cycle = temp_cycle
    return tuple(cycle) , weight

def matching_christofides_search(graph) :
    #1.find the list of edges representing a minimum weight perfect matching .
    for edge in graph.edges() :
        graph[edge[0]][edge[1]]['weight'] = -1 * graph[edge[0]][edge[1]]['weight']
    minimum_weight_perfect_matching = list(max_weight_matching(graph, maxcardinality=True))
    for edge in graph.edges() :
        graph[edge[0]][edge[1]]['weight'] = -1 * graph[edge[0]][edge[1]]['weight']
    #2.find the cycle that represents the solution for christofides algorithm .
    cycle , weight = christofides(graph)
    cycle = list(cycle)
    new_cycle = cycle.copy()
    #3.change some edges trying get better solution .
    for edge in minimum_weight_perfect_matching :
        index1 = min(cycle.index(edge[0]),cycle.index(edge[1]))
        index2 = max(cycle.index(edge[0]),cycle.index(edge[1]))
        new_cycle[index1+1:index2+1] = list(reversed(new_cycle[index1+1:index2+1]))
        new_weight = compute_cycle_length(graph,new_cycle)
        if(int(new_weight) < int(weight)) :
            cycle = new_cycle
            weight = new_weight
    return tuple(cycle) , weight
```

```python
#This cell is for the case proposed to be added to the project .
#Dynamic programming and Christofides with incomplete graphs .
def load_graph_incomplete(filename) :
    """
    This functions is used to create graph instance from specific file with specific format .
    File foramt :
    vertex vertex weight
    .
    .
    .
    .
    .
    EOF
    """
    file = open(filename,'r')
    lines = file.readlines()
    graph = nx.Graph()
    for line in lines :
        line = line.strip().split(" ")
        if (line[0] == 'EOF') :
            break
        graph.add_edge(int(line[0]),int(line[1]),weight = int(line[2]))
    for i in range(graph.number_of_nodes()) :
        for j in range(i) :
            if((i,j) not in graph.edges()) :
                graph.add_edge(i,j,weight = float('inf'))
    return graph

def plot_graph_incomplete(graph) :
    """
    This function takes a graph instance and draw it .
    Note : Here positions of vertices is not considered .
    """
    temp_graph = nx.Graph()
    for edge in graph.edges() :
        if(graph[edge[0]][edge[1]]['weight'] != float('inf')) :
            temp_graph.add_edge(edge[0],edge[1],weight = graph[edge[0]][edge[1]]['weight'])
    pos = nx.spring_layout(temp_graph)
    nx.draw_networkx_nodes(temp_graph,pos)
    nx.draw_networkx_edges(temp_graph,pos)
    nx.draw_networkx_labels(temp_graph,pos)
    #nx.draw_networkx_edge_labels(graph,pos)
    plt.show()
    
    
def dynamic_programming_incomplete(graph) :
    """This function takes only one parameter
    graph : networkx's instance .
    This is the implementation of Dynamic Programming approach.
    This function returns the minimum weight Hamiltonian cycle wth its weight .
    """
    #1.We must create a list that contains all possible subsets of our {all vertices without zero }set .
    number_of_vertices = graph.number_of_nodes()
    chain_of_subsets = chain.from_iterable(combinations(range(1,number_of_vertices),r) for r in range(number_of_vertices+1))
    #2.Handle the case of subsets with only one element ---> this means edges from 0 to each other vertex .
    weight_of_subset = dict()
    for vertex in range(1,number_of_vertices) :
        weight_of_subset[(vertex,),vertex] = graph[0][vertex]['weight']
    #3.Handling the case of subsets that have more than one element .
    for subset in chain_of_subsets :
        if(len(subset) > 1) :
            for vertex in subset :
                sub_subset = [node for node in subset if node != vertex]
                sub_subset = tuple(sub_subset)
                weight_of_subset[subset,vertex] = min([weight_of_subset[sub_subset,extension] + graph[extension][vertex]['weight'] for extension in sub_subset])
    #4.ensuring whether there is a solution or not .
    decision = False
    for vertex in range(1,number_of_vertices) :
        if(weight_of_subset[tuple(range(1,number_of_vertices)),vertex] + graph[vertex][0]['weight'] != float('inf')) :
            decision = True
            break
    if(decision == False) :
        return ('Nan','Nan')
    #5.Preparing the results to be returned .
    min_weight = float("inf")
    cycle = tuple()
    new_last_vertex = None
    for vertex in range(1,number_of_vertices) :
        if(weight_of_subset[tuple(range(1,number_of_vertices)),vertex] + graph[vertex][0]['weight'] < min_weight) :
            min_weight = weight_of_subset[tuple(range(1,number_of_vertices)),vertex] + graph[vertex][0]['weight']
            new_last_vertex = vertex
    new_weight = min_weight  - graph[new_last_vertex][0]['weight']
    path = list()
    path.append(0)
    path.append(new_last_vertex)
    for i in reversed(range(1,number_of_vertices-1)) :
        list_of_combinations = list(combinations(range(1,number_of_vertices),i))
        for j in range(number_of_vertices) :
            for sub in list_of_combinations :
                if(j in sub and set(path).isdisjoint(sub)) :
                    if(weight_of_subset[sub,j] + graph[j][new_last_vertex]['weight'] - new_weight < 0.0000000001) :
                        new_weight = new_weight - graph[j][new_last_vertex]['weight']
                        new_last_vertex = j
                        path.append(j)
    if(min_weight == float('inf')) :
        return ("Nan","Nan")
    else :
        return tuple(path) , min_weight

def christofides_incomplete(graph,source_vertex = 0) :
    """
    This function is an implementation of Christofides algorithm .
    graph : networkx instance .
    source_vertex : The starting and ending point of the cycle .
    """
    #1.compute MST .
    mst = nx.minimum_spanning_tree(graph)
    #2.find the vertices with odd degrees .
    odd_degree_vertices = list()
    for vertex in graph.nodes() :
        if(mst.degree(vertex) % 2 != 0) :
            odd_degree_vertices.append(vertex)
    #3.build subgraph with only vertices with odd degrees and the edges between them .
    odd_vertices_graph = graph.subgraph(odd_degree_vertices)
    #4.compute minimum weight perfect matching on the odd vertices subgraph .
    for edge in odd_vertices_graph.edges() :
        odd_vertices_graph[edge[0]][edge[1]]['weight'] = -1 * odd_vertices_graph[edge[0]][edge[1]]['weight']
    minimum_weight_perfect_matching = list(max_weight_matching(odd_vertices_graph, maxcardinality=True))
    #To correct negative edges .
    for edge in graph.edges() :
        graph[edge[0]][edge[1]]['weight'] = abs(graph[edge[0]][edge[1]]['weight'])
    #5.combine the mst with the edges from minimum weight perfect matching .
    multi_graph = nx.MultiGraph(mst)
    for edge in minimum_weight_perfect_matching :
        multi_graph.add_edge(edge[0],edge[1],weight = graph[edge[0]][edge[1]]['weight'])
    #6.find Euler's cycle
    euler_cycle = list(eulerian_circuit(multi_graph,source = source_vertex))
    cycle = list()
    for edge in euler_cycle :
        if(edge[0] not in cycle) :
            cycle.append(edge[0])
        if(edge[1] not in cycle) :
            cycle.append(edge[1])
    weight = compute_cycle_length(graph,tuple(cycle))
    if(weight == float('inf')) :
        return ("Nan","Nan")
    else :
        return tuple(cycle) , weight
```

```python
#This cell is written to discuss the number of expanded vertices in the branch and bound approach .
#Note : comparision between performance of branch and bound in tight VS not tight examples . 

def compute_lower_bound(graph,cycle) :
    """This function takes a graph instance and a subcycle and returns the lower bound in this stage.  """
    """graph : networkx instance"""
    """cycle : tuple and in some cases can be passed as a list. """
    """This function is used in the function that implements branch and bound algorithm.  """
    #1.computing the length of the path of visited nodes .
    weight_of_visited_vertices = 0
    for i in range(len(cycle) - 1 ) :
        weight_of_visited_vertices += graph[cycle[i]][cycle[i+1]]['weight']
    #2.computing the remaining graph .
    unvisited = [vertex for vertex in graph.nodes() if vertex not in cycle]
    remaining_graph = graph.subgraph(unvisited)
    #3.computing the weight of MST in the remaining graph .
    mst_edges = nx.minimum_spanning_edges(remaining_graph)
    mst_weight = 0 
    for edge in mst_edges :
        mst_weight += edge[2]['weight']
    #4.Defining the behaviour when we reach the end of the cycle .
    #Note : In this implementation the length of cycle can not be equal to zero .
    #Case when all vertices either in cycle or in the remaining graph .
    if (len(cycle) == 0 or len(cycle) == graph.number_of_nodes()) :
        return mst_weight + weight_of_visited_vertices
    #5.handling the case when some vertices belong to the remaining graph and others belong to the cycle .
    first_vertex_in_cycle = cycle[0]
    last_vertex_in_cycle = cycle[-1]
    min_weight_to_first_vertex = float("inf")
    min_weight_from_last_vertex = float("inf")
    for vertex in graph.nodes() :
        if vertex not in cycle :
            if(graph[vertex][first_vertex_in_cycle]['weight'] < min_weight_to_first_vertex) :
                min_weight_to_first_vertex = graph[vertex][first_vertex_in_cycle]['weight']
    for vertex in graph.nodes() :
        if vertex not in cycle :
            if(graph[last_vertex_in_cycle][vertex]['weight'] < min_weight_from_last_vertex) :
                min_weight_from_last_vertex = graph[last_vertex_in_cycle][vertex]['weight']
    return weight_of_visited_vertices + mst_weight + min_weight_to_first_vertex + min_weight_from_last_vertex

def branch_and_bound_helper(graph,list_of_cycles,number_of_expanded_nodes,cycle=None,current_min = float("inf")) :
    """
    This function takes three parameters :
    graph : networkx's instance in which we want to find a minimum weight Hamiltonian cycle .
    cycle : The vertex that is a part of the cycle under cosideration at each stage (tuple) .
    current_min : The weight of the best solution at each stage .
    This function is an implementation of brunch and bound algoritm .
    This function returns the minimum weight Hamitonian cycle and its weight .
    
    """
    #1.If there is no root . we define the root of search tree .
    if cycle is None :
        cycle = [0]
    #2.The case when we reach the end of the cycle and we're ready to compute its weight.
    if (len(cycle) == graph.number_of_nodes()) :
        weight = sum([graph[cycle[i]][cycle[(i + 1) % len(cycle)]]['weight'] for i in range(len(cycle))])
        return weight
    #3.Determining the unvisited vertices .
    unvisited = []
    for vertex in graph.nodes():
        if vertex not in cycle:
            unvisited.append((graph[cycle[-1]][vertex]['weight'], vertex))
    #4.Fininding nearest neighbors to current last vertex of the current part of a cycle .
    unvisited = sorted(unvisited)
    #5.Trying reach the end of the cycle .
    for (weight,vertex) in unvisited :
        #Tuples doesn't support append .
        modified_cycle = list(cycle)
        modified_cycle.append(vertex)
        #To check if we by adding this vertex to the current path(part of cycle) able to reach a better solution
        #better solution -----> There is a possiblity to find smaller weight cycle by proceeding in this path .
        if compute_lower_bound(graph,modified_cycle) < current_min :
            number_of_expanded_nodes[0] = number_of_expanded_nodes[0] + 1
            new_weight = branch_and_bound_helper(graph,list_of_cycles,number_of_expanded_nodes,modified_cycle,current_min)
            #If The newly found cycle is better than the previously discovered then save it .
            if new_weight < current_min :
                current_min = new_weight
                if(len(modified_cycle) == graph.number_of_nodes()) :
                    list_of_cycles.append(modified_cycle)
    return current_min

def branch_and_bound(graph) :
    list_of_cycles = list()
    number_of_expanded_nodes = list([1])
    weight_of_best_hamiltonian_cycle = branch_and_bound_helper(graph,list_of_cycles,number_of_expanded_nodes)
    return tuple(list_of_cycles[-1]) , weight_of_best_hamiltonian_cycle , int(number_of_expanded_nodes[-1])
```

```python
#This cell is representing the steps of executing double approximation algorithm on one specific graph
graph = load_graph('C:/Users/Ayman/datasetsLoadGraph/paperExample.txt')
decision = False
if (is_valid(graph)) :
    decision = True
if(decision) :
    print("Is the loaded graph valid?\nThe answer is yes . \n ")
else :
    print("Is the loaded graph valid?\nThe answer is no . \n ")
    
def christofides(graph,source_vertex = 0) :
    """
    This function is an implementation of Christofides algorithm .
    graph : networkx instance .
    source_vertex : The starting and ending point of the cycle .
    """
    #1.compute MST .
    print("1.Computing minimum weight spanning tree .")
    mst = nx.minimum_spanning_tree(graph)
    print("The mst edges are : "+str(mst.edges()))
    #2.find the vertices with odd degrees .
    print("2.Find the vertices with odd degrees .")
    odd_degree_vertices = list()
    for vertex in graph.nodes() :
        if(mst.degree(vertex) % 2 != 0) :
            odd_degree_vertices.append(vertex)
    for vertex in odd_degree_vertices :
        print("The vertex " +str(vertex) + " has odd degree " + str(graph.degree(vertex))+ " .")
    #3.build subgraph with only vertices with odd degrees and the edges between them .
    odd_vertices_graph = graph.subgraph(odd_degree_vertices)
    #4.compute minimum weight perfect matching on the odd vertices subgraph .
    for edge in odd_vertices_graph.edges() :
        odd_vertices_graph[edge[0]][edge[1]]['weight'] = -1 * odd_vertices_graph[edge[0]][edge[1]]['weight']
    minimum_weight_perfect_matching = list(max_weight_matching(odd_vertices_graph, maxcardinality=True))
    print("3.Compute the minimum weight perfect matching between odd degree vertices .")
    print("Edges resulted by matching " + str(minimum_weight_perfect_matching))
    #To correct negative edges .
    for edge in graph.edges() :
        graph[edge[0]][edge[1]]['weight'] = abs(graph[edge[0]][edge[1]]['weight'])
    #5.combine the mst with the edges from minimum weight perfect matching .
    multi_graph = nx.MultiGraph(mst)
    for edge in minimum_weight_perfect_matching :
        multi_graph.add_edge(edge[0],edge[1],weight = graph[edge[0]][edge[1]]['weight'])
    print("4.Combining mst with edges from matching .")
    print("Resulted graph edges are "+str(multi_graph.edges))
    #6.find Euler's cycle
    euler_cycle = list(eulerian_circuit(multi_graph,source = source_vertex))
    print("5.Find Euler's cycle .")
    print(euler_cycle)
    cycle = list()
    for edge in euler_cycle :
        if(edge[0] not in cycle) :
            cycle.append(edge[0])
        if(edge[1] not in cycle) :
            cycle.append(edge[1])
    weight = compute_cycle_length(graph,tuple(cycle))
    return tuple(cycle) , weight

sol = christofides(graph)
print("The result is ")
print("The cycle is "+str(sol[0]))
print("The cycle's weight is "+str(sol[1]))
```

```python
#This cell is the recoding of function that decides wether a graph is valid or not .
def is_triangle_inequal(graph) :
    nodes = graph.nodes()
    n = graph.number_of_nodes()
    triangle_inequality = True
    for v1 in range(n) :
        for v2 in range(n) :
            for v3 in range(n) :
                if(v1 != v2 and v1 != v3 and v2 != v3) :
                    if ((graph[v1][v3]['weight'] + graph[v3][v2]['weight']) < graph[v1][v2]['weight'] and abs(graph[v1][v3]['weight'] + graph[v3][v2]['weight'] - graph[v1][v2]['weight']) < 0.00000001 ) :
                        triangle_inequality = False 
                        break
    return triangle_inequality
def is_valid(graph) :
    decision = False
    print("Undirected " + str(not graph.is_directed()))
    print("Simple " + str(not graph.is_multigraph()))
    if(not graph.is_directed() and not graph.is_multigraph()) :
        n = graph.number_of_nodes()
        e = graph.number_of_edges()
        print("Complete " + str(((n-1)*n) / 2 == e))
        if( ((n-1)*n) / 2 == e ) :
            print("Triangle " +str(is_triangle_inequal(graph)))
            if(is_triangle_inequal(graph)) :
                decision = True
    return decision
```

```python
#This cell contains all the main functions .
#This cell only for complete graphs .
#Later cell will be used to discuss the case of  incomplete graphs with christofides and dynamic programming .
import networkx as nx
import os
import matplotlib.pyplot as plt
import math
from itertools import permutations
from itertools import combinations
from itertools import chain
import numpy as np
from networkx.algorithms.matching import max_weight_matching
from networkx.algorithms.euler import eulerian_circuit
import pandas as pd
import time

def load_coordinates_from_file(filename) :
    """This function takes the name of the file that containes the coordinates and returns list of points.  """
    """filename : String"""
    with open(filename,'r') as file  :
        list_of_points = [] 
        points = file.readlines()
        for point in points :
            point = point.strip().split(" ")
            if(point[0] == 'EOF') :
                return list_of_points
            x_coordinate = point[1]
            y_coordinate = point[2]
            list_of_points.append((x_coordinate,y_coordinate))
            
def load_graph(filename) :
    """
    This functions is used to create graph instance from specific file with specific format .
    File foramt :
    vertex vertex weight
    .
    .
    .
    .
    .
    EOF
    """
    file = open(filename,'r')
    lines = file.readlines()
    graph = nx.Graph()
    for line in lines :
        line = line.strip().split(" ")
        if (line[0] == 'EOF') :
            return graph
        graph.add_edge(int(line[0]),int(line[1]),weight = int(line[2]))
        
def generate_tight_example(number_of_vertices,file = False) :
    """
    This function is used to generate graph instance which has a special case that called tight example .
    If the file parameter is True ,this means generate file representing tight example  .
    """
    graph = nx.Graph()
    for i in range(number_of_vertices) :
        for j in range(i) :
            if(j==0) :
                graph.add_edge(i,j,weight = 1)
            else :
                graph.add_edge(i,j,weight = 2)
    if(file == True) :
        file = open("C:/Users/Ayman/datasets/tightExample"+str(graph.number_of_nodes())+".txt",'w')
        for edge in graph.edges() :
            file.write(str(edge[0])+" "+str(edge[1])+" "+str(graph[edge[0]][edge[1]]['weight'])+"\n")
        file.write('EOF')
        file.close()
    return graph
        
def compute_distance(point1,point2) :
    """this function takes two points,each of them is 2-tuple and returns a Euclidean distance between them.  """
    """point1 : 2-tuple"""
    """point2 : 2-tuple"""
    x1 = float(point1[0])
    y1 = float(point1[1])
    x2 = float(point2[0])
    y2 = float(point2[1])
    #distance  = int(math.sqrt((x1 - x2) **2 + (y1 - y2) **2) + 0.5)   --makes the graph sometimes not metric .
    distance  = math.sqrt((x1 - x2) **2 + (y1 - y2) **2)
    return distance

def convert_coordinates_to_graph(list_of_points) :
    """This function takes a list of 2-tuples and returns a graph instance of networkx' class.  """
    """list_of_points : list of 2-tuples"""
    graph = nx.Graph()
    number_of_vertices = len(list_of_points)
    for i in range(number_of_vertices) :
        for j in range(i) :
            graph.add_edge(i,j,weight = compute_distance(list_of_points[i],list_of_points[j]))
    return graph

def plot_cycle(list_of_points, cycle):
    """This function takes a list_of_points that represents the coordinates of the points and a cycle which is a list 
    and draw the cycle.    """
    """list_of_points : list of 2-tuples"""
    x_coordinates = [float(list_of_points[i][0]) for i in cycle]
    y_coordinates = [float(list_of_points[i][1]) for i in cycle]
    x_coordinates.append(float(list_of_points[cycle[0]][0]))
    y_coordinates.append(float(list_of_points[cycle[0]][1]))
    plt.plot(x_coordinates, y_coordinates, 'or-', )
    plt.show()

def plot_graph(graph) :
    """
    This function takes a graph instance and draw it .
    Note : Here positions of vertices is not considered .
    """
    pos = nx.circular_layout(graph)
    nx.draw_networkx_nodes(graph,pos)
    nx.draw_networkx_edges(graph,pos)
    nx.draw_networkx_labels(graph,pos)
    labels = nx.get_edge_attributes(graph,'weight')
    nx.draw_networkx_edge_labels(graph,pos,edge_labels=labels)
    plt.show()
    

def draw_graph(filename) :
    """
    This function takes positions of the vertices into account .
    """
    list_of_points = load_coordinates_from_file(filename)
    graph = nx.Graph()
    for index , point in enumerate(list_of_points) :
        graph.add_node(index,pos = (float(point[0]),float(point[1])))
    for i in range(len(list_of_points)) :
        for j in range(i) :
            graph.add_edge(i,j,weight = int(compute_distance(list_of_points[i],list_of_points[j])))
    pos = nx.get_node_attributes(graph,'pos')
    nx.draw_networkx_nodes(graph,pos)
    nx.draw_networkx_edges(graph,pos)
    nx.draw_networkx_labels(graph,pos)
    
def compute_cycle_length(graph,cycle) :
    """This function takes a graph instance and a cycle and returns the length of this cycle.  """
    """graph : networkx instance"""
    """cycle : tuple and in some cases can be passed as a list. """
    length = 0
    for vertex_id in range(len(cycle)) :
        length += graph[cycle[vertex_id]][cycle[(vertex_id + 1) % len(cycle)]]['weight']
    return length
  
    
def brute_force(graph) :
    """This function takes a graph instance and computes a minimum weigth Hamiltonian cycle on it.  """
    """graph : networkx instance"""
    number_of_vertices = graph.number_of_nodes()
    minimum_weight = float("+inf")
    minimum_weight_hamiltonian_cycle = None
    #1.iterate through all permutations and choose the best cycle (here : cycle with the minimum length)
    for cycle in permutations(range(number_of_vertices)) :
        #This line is added in 2/6/2020 10:08 pm
        #The reason is to reduce the number of candidate cycles from n! to (n-1)!
        if(cycle[0] != 0 ) :
            return minimum_weight_hamiltonian_cycle , minimum_weight
        current_weight = compute_cycle_length(graph,cycle)
        if(current_weight < minimum_weight ) :
            minimum_weight = current_weight
            minimum_weight_hamiltonian_cycle = cycle
    return minimum_weight_hamiltonian_cycle , minimum_weight
            

def compute_lower_bound(graph,cycle) :
    """This function takes a graph instance and a subcycle and returns the lower bound in this stage.  """
    """graph : networkx instance"""
    """cycle : tuple and in some cases can be passed as a list. """
    """This function is used in the function that implements branch and bound algorithm.  """
    #1.computing the length of the path of visited nodes .
    weight_of_visited_vertices = 0
    for i in range(len(cycle) - 1 ) :
        weight_of_visited_vertices += graph[cycle[i]][cycle[i+1]]['weight']
    #2.computing the remaining graph .
    unvisited = [vertex for vertex in graph.nodes() if vertex not in cycle]
    remaining_graph = graph.subgraph(unvisited)
    #3.computing the weight of MST in the remaining graph .
    mst_edges = nx.minimum_spanning_edges(remaining_graph)
    mst_weight = 0 
    for edge in mst_edges :
        mst_weight += edge[2]['weight']
    #4.Defining the behaviour when we reach the end of the cycle .
    #Note : In this implementation the length of cycle can not be equal to zero .
    #Case when all vertices either in cycle or in the remaining graph .
    if (len(cycle) == 0 or len(cycle) == graph.number_of_nodes()) :
        return mst_weight + weight_of_visited_vertices
    #5.handling the case when some vertices belong to the remaining graph and others belong to the cycle .
    first_vertex_in_cycle = cycle[0]
    last_vertex_in_cycle = cycle[-1]
    min_weight_to_first_vertex = float("inf")
    min_weight_from_last_vertex = float("inf")
    for vertex in graph.nodes() :
        if vertex not in cycle :
            if(graph[vertex][first_vertex_in_cycle]['weight'] < min_weight_to_first_vertex) :
                min_weight_to_first_vertex = graph[vertex][first_vertex_in_cycle]['weight']
    for vertex in graph.nodes() :
        if vertex not in cycle :
            if(graph[last_vertex_in_cycle][vertex]['weight'] < min_weight_from_last_vertex) :
                min_weight_from_last_vertex = graph[last_vertex_in_cycle][vertex]['weight']
    return weight_of_visited_vertices + mst_weight + min_weight_to_first_vertex + min_weight_from_last_vertex

def branch_and_bound_helper(graph,list_of_cycles,cycle=None,current_min = float("inf") ) :
    """
    This function takes three parameters :
    graph : networkx's instance in which we want to find a minimum weight Hamiltonian cycle .
    cycle : The vertex that is a part of the cycle under cosideration at each stage (tuple) .
    current_min : The weight of the best solution at each stage .
    This function is an implementation of brunch and bound algoritm .
    This function returns the minimum weight Hamitonian cycle and its weight .
    
    """
    #1.If there is no root . we define the root of search tree .
    if cycle is None :
        cycle = [0]
    #2.The case when we reach the end of the cycle and we're ready to compute its weight.
    if (len(cycle) == graph.number_of_nodes()) :
        weight = sum([graph[cycle[i]][cycle[(i + 1) % len(cycle)]]['weight'] for i in range(len(cycle))])
        return weight
    #3.Determining the unvisited vertices .
    unvisited = []
    for vertex in graph.nodes():
        if vertex not in cycle:
            unvisited.append((graph[cycle[-1]][vertex]['weight'], vertex))
    #4.Fininding nearest neighbors to current last vertex of the current part of a cycle .
    unvisited = sorted(unvisited)
    #5.Trying reach the end of the cycle .
    for (weight,vertex) in unvisited :
        #Tuples doesn't support append .
        modified_cycle = list(cycle)
        modified_cycle.append(vertex)
        #To check if we by adding this vertex to the current path(part of cycle) able to reach a better solution
        #better solution -----> There is a possiblity to find smaller weight cycle by proceeding in this path .
        if compute_lower_bound(graph,modified_cycle) < current_min :
            new_weight = branch_and_bound_helper(graph,list_of_cycles,modified_cycle,current_min)
            #If The newly found cycle is better than the previously discovered then save it .
            if new_weight < current_min :
                current_min = new_weight
                if(len(modified_cycle) == graph.number_of_nodes()) :
                    list_of_cycles.append(modified_cycle)
    return current_min

def branch_and_bound(graph) :
    list_of_cycles = list()
    weight_of_best_hamiltonian_cycle = branch_and_bound_helper(graph,list_of_cycles)
    return tuple(list_of_cycles[-1]) , weight_of_best_hamiltonian_cycle

def dynamic_programming(graph) :
    """This function takes only one parameter
    graph : networkx's instance .
    This is the implementation of Dynamic Programming approach.
    This function returns the minimum weight Hamiltonian cycle wth its weight .
    """
    #1.We must create a list that contains all possible subsets of our {all vertices without zero }set .
    n = graph.number_of_nodes()
    chain_of_subsets = chain.from_iterable(combinations(range(1,n),r) for r in range(n+1))
    #2.Handle the case of subsets with only one element ---> this means edges from 0 to each other vertex .
    weight_of_subset = dict()
    for vertex in range(1,n) :
        weight_of_subset[(vertex,),vertex] = graph[0][vertex]['weight']
    #3.Handling the case of subsets that have more than one element .
    for subset in chain_of_subsets :
        if(len(subset) > 1) :
            for vertex in subset :
                sub_subset = [node for node in subset if node != vertex]
                sub_subset = tuple(sub_subset)
                weight_of_subset[subset,vertex] = min([weight_of_subset[sub_subset,extension] + graph[extension][vertex]['weight'] for extension in sub_subset])
    #4.Preparing the results to be returned .
    min_weight = float("inf")
    cycle = tuple()
    new_last_vertex = None
    for vertex in range(1,n) :
        if(weight_of_subset[tuple(range(1,n)),vertex] + graph[vertex][0]['weight'] < min_weight) :
            min_weight = weight_of_subset[tuple(range(1,n)),vertex] + graph[vertex][0]['weight']
            new_last_vertex = vertex
    new_weight = min_weight  - graph[new_last_vertex][0]['weight']
    path = list()
    path.append(0)
    path.append(new_last_vertex)
    for i in reversed(range(1,n-1)) :
        list_of_combinations = list(combinations(range(1,n),i))
        for j in range(n) :
            for sub in list_of_combinations :
                if(j in sub and set(path).isdisjoint(sub)) :
                    if(weight_of_subset[sub,j] + graph[j][new_last_vertex]['weight'] - new_weight < 0.0000000001) :
                        new_weight = new_weight - graph[j][new_last_vertex]['weight']
                        new_last_vertex = j
                        path.append(j)
    return path , min_weight

def double_approximation(graph,source_vertex = 0) :
    """This function implements the double approximation algorithm to solve Metric TSP."""
    #1.Compute the MST for the inputed graph .
    mst = nx.minimum_spanning_tree(graph) 
    #2.Traverse the mst .
    cycle = nx.dfs_preorder_nodes(mst,source_vertex)
    cycle = tuple(cycle)
    weight = compute_cycle_length(graph,cycle)
    return cycle , weight

def modified_double_approximation(graph) :
    """
    This function is cosidered as a modification for double approximation function .
    The mian point here is that the order of visiting the vertices is important because it plays a major role in identifying
    the bypasses .
    """
    min_weight = float("inf")
    min_cycle = None 
    n = graph.number_of_nodes()
    for vertex in range(n) :
        cycle , weight = double_approximation(graph,vertex)
        if(weight < min_weight) :
            min_weight = weight
            min_cycle = cycle
    return min_cycle , min_weight

def random_permutation(graph,threshold = 2,max_time = float("inf") ) :
    tic = time.time()
    weight = nx.minimum_spanning_tree(graph).size(weight = 'weight')
    min_weight = float("inf")
    min_cycle = None
    number_of_vertices = graph.number_of_nodes()
    toc = time.time()
    while(round((toc - tic) * 1000) < max_time and min_weight > threshold * weight) :
        cycle = np.random.permutation(number_of_vertices)
        min_weight = compute_cycle_length(graph,cycle)
        toc = time.time()
    if(cycle is None) :
        return None , None
    else :
        return tuple(cycle) , min_weight
    
def nearest_neighbor(graph,source_vertex = 0) :
    """
    This function is the implementation of choosing the nearest vertex(neighbor) each time .
    graph : networkx's instance .
    source_vertex : The first vertex in the cycle .
    """
    #1.Intialization step for neccessary variables .
    number_of_vertices = graph.number_of_nodes()
    cycle = list([source_vertex])
    last_visited_vertex = source_vertex
    unvisited = list([vertex for vertex in range(number_of_vertices) if vertex != source_vertex])
    #2.Passing throught all unvisited vertices .
    for counter in range(number_of_vertices - 1):
        next_visited_vertex = None
        weight = float('inf')
        for vertex in unvisited :
            if(graph[last_visited_vertex][vertex]['weight'] < weight) :
                next_visited_vertex = vertex
                weight = graph[last_visited_vertex][vertex]['weight']
        cycle.append(next_visited_vertex)
        unvisited.remove(next_visited_vertex)
        last_visited_vertex = next_visited_vertex
    weight = compute_cycle_length(graph,cycle)
    return tuple(cycle) , weight

def modified_nearest_neighbor(graph) :
    """
    This function is slight modification of nearest neighbor ,considering that the starting vertex can affect the result .
    Main point : Order is important (Each vertex affects the choosing of next one. )
    """
    number_of_vertices = graph.number_of_nodes()
    cycle = None
    weight = float("inf")
    for vertex in range(number_of_vertices) :
        new_cycle , new_weight = nearest_neighbor(graph,vertex)
        if (new_weight < weight) :
            weight = new_weight
            cycle = new_cycle
    return cycle , weight
    
def christofides(graph,source_vertex = 0) :
    """
    This function is an implementation of Christofides algorithm .
    graph : networkx instance .
    source_vertex : The starting and ending point of the cycle .
    """
    #1.compute MST .
    mst = nx.minimum_spanning_tree(graph)
    #2.find the vertices with odd degrees .
    odd_degree_vertices = list()
    for vertex in graph.nodes() :
        if(mst.degree(vertex) % 2 != 0) :
            odd_degree_vertices.append(vertex)
    #3.build subgraph with only vertices with odd degrees and the edges between them .
    odd_vertices_graph = graph.subgraph(odd_degree_vertices)
    #4.compute minimum weight perfect matching on the odd vertices subgraph .
    for edge in odd_vertices_graph.edges() :
        odd_vertices_graph[edge[0]][edge[1]]['weight'] = -1 * odd_vertices_graph[edge[0]][edge[1]]['weight']
    minimum_weight_perfect_matching = list(max_weight_matching(odd_vertices_graph, maxcardinality=True))
    #To correct negative edges .
    for edge in graph.edges() :
        graph[edge[0]][edge[1]]['weight'] = abs(graph[edge[0]][edge[1]]['weight'])
    #5.combine the mst with the edges from minimum weight perfect matching .
    multi_graph = nx.MultiGraph(mst)
    for edge in minimum_weight_perfect_matching :
        multi_graph.add_edge(edge[0],edge[1],weight = graph[edge[0]][edge[1]]['weight'])
    #6.find Euler's cycle
    euler_cycle = list(eulerian_circuit(multi_graph,source = source_vertex))
    cycle = list()
    for edge in euler_cycle :
        if(edge[0] not in cycle) :
            cycle.append(edge[0])
        if(edge[1] not in cycle) :
            cycle.append(edge[1])
    weight = compute_cycle_length(graph,tuple(cycle))
    return tuple(cycle) , weight

def modified_christofides(graph) :
    """
    This function is used to negotiate the effects of visiting order .
    """
    min_weight = float("inf")
    min_cycle = None
    for vertex in range(graph.number_of_nodes()) :
        cycle , weight = christofides(graph,vertex)
        if (weight < min_weight ) :
            min_weight = weight
            min_cycle = cycle
    return min_cycle , min_weight
      
def first_choice_hill_climbing(graph,ratio = 100) :
    """
    This functions is an implementatioan of local search algorithm called first choice hill climbing .
    There are many reasons that lead to choose this version among other hill climbing versions .
    Note : Here we consider only change between two edges .
    """
    #1.produce a random permutation .
    number_of_vertices = graph.number_of_nodes()
    cycle = np.random.permutation(number_of_vertices)
    counter = np.random.randint(0,10)
    for i in range(counter) :
        np.random.shuffle(cycle)
    weight = compute_cycle_length(graph,cycle)
    #2.change the positions of two randomly choosen vertices in the cycle if the condition holds .
    for iteration in range(ratio * number_of_vertices) :
        num1 = np.random.randint(1,number_of_vertices)
        num2 = np.random.randint(1,number_of_vertices)
        while(num1 == num2) :
            num1 = np.random.randint(1,number_of_vertices)
            num2 = np.random.randint(1,number_of_vertices)
        first_index = min(num1,num2)
        second_index = max(num1,num2)
        temp_cycle = cycle.copy()
        temp_cycle[first_index:second_index + 1] = list(reversed(temp_cycle[first_index:second_index + 1]))
        new_weight = compute_cycle_length(graph,temp_cycle)
        if(new_weight < weight) :
            weight = new_weight
            cycle = temp_cycle
    return tuple(cycle) , weight 

def christofides_first_choice_hill_climbing(graph,ratio = 100) :
    """
    This is a modification of first choice hill climbing .
    instead of starting with a random cycle , we start with a cycle given as a solution by christofides algorithm .
    """
    #1.produce a christofides solution .
    number_of_vertices = graph.number_of_nodes()
    cycle , weight = christofides(graph)
    cycle = list(cycle)
    #2.change the positions of two randomly choosen vertices in the cycle if the condition holds .
    for iteration in range(ratio * number_of_vertices) :
        num1 = np.random.randint(1,number_of_vertices)
        num2 = np.random.randint(1,number_of_vertices)
        while(num1 == num2) :
            num1 = np.random.randint(1,number_of_vertices)
            num2 = np.random.randint(1,number_of_vertices)
        first_index = min(num1,num2)
        second_index = max(num1,num2)
        temp_cycle = cycle.copy()
        temp_cycle[first_index:second_index + 1] = list(reversed(temp_cycle[first_index:second_index + 1]))
        new_weight = compute_cycle_length(graph,temp_cycle)
        if(new_weight < weight) :
            weight = new_weight
            cycle = temp_cycle
    return tuple(cycle) , weight

def matching_christofides_search(graph) :
    #1.find the list of edges representing a minimum weight perfect matching .
    for edge in graph.edges() :
        graph[edge[0]][edge[1]]['weight'] = -1 * graph[edge[0]][edge[1]]['weight']
    minimum_weight_perfect_matching = list(max_weight_matching(graph, maxcardinality=True))
    for edge in graph.edges() :
        graph[edge[0]][edge[1]]['weight'] = -1 * graph[edge[0]][edge[1]]['weight']
    #2.find the cycle that represents the solution for christofides algorithm .
    cycle , weight = christofides(graph)
    cycle = list(cycle)
    new_cycle = cycle.copy()
    #3.change some edges trying get better solution .
    for edge in minimum_weight_perfect_matching :
        index1 = min(cycle.index(edge[0]),cycle.index(edge[1]))
        index2 = max(cycle.index(edge[0]),cycle.index(edge[1]))
        new_cycle[index1+1:index2+1] = list(reversed(new_cycle[index1+1:index2+1]))
        new_weight = compute_cycle_length(graph,new_cycle)
        if(int(new_weight) < int(weight)) :
            cycle = new_cycle
            weight = new_weight
    return tuple(cycle) , weight
```
